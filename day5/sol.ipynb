```python
seeds = list(map(int, data[0].replace('seeds: ', '').split()))
seeds
```

```python
maps = []
for i in range(1, len(data)):
  map_str_data = data[i].splitlines()[1:]
  map_data = [];
  for row in map_str_data:
    map_data.append(list(map(int, row.split())))
  maps.append(map_data)

maps
```

```python
print(range)
ranges = [(seeds[i], seeds[i] + seeds[i + 1]) for i in range(0, len(seeds) - 1, 2)]
ranges

for mapping in maps:
  future_ranges = []
  for row in mapping:
    unfiltered_ranges = []
    row_s = row[1]
    row_e = row[1] + row[2]
    for range2 in ranges:
      overlap_s = max(row_s, range2[0])
      overlap_e = min(row_e, range2[1])
      if overlap_s < overlap_e:
        if range2[0] < overlap_s:
          unfiltered_ranges.append((range2[0], overlap_s))
        if overlap_e < range2[1]:
          unfiltered_ranges.append((overlap_e, range2[1]))
        diff = overlap_s - row[1]
        future_ranges.append((row[0] + diff, row[0] + diff + (overlap_e - overlap_s)))
      else:
        unfiltered_ranges.append(range2)
    ranges = unfiltered_ranges
  ranges = future_ranges + ranges

min(map(lambda x: x[0], ranges))
```
