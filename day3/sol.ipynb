```python
with open('input.txt') as f:
  matrix = f.read().splitlines()

# str_matrix = '''467..114..
# ...*......
# ..35..633.
# ......#...
# 617*......
# .....+.58.
# ..592.....
# ......755.
# ...$.*....
# .664.598..'''

# matrix = str_matrix.splitlines()
matrix
```

```python
def is_symbol(x: str):
  return x != '.' and not x.isnumeric()

nums = []

for i in range(len(matrix)):
  has_symbol = False
  num_start = num_end = None

  def update_final(has_symbol):
    if num_start == None:
      return
    if num_end + 1 < len(matrix[i]):
      has_symbol |= is_symbol(matrix[i][num_end + 1])
      if i > 0:
        has_symbol |= is_symbol(matrix[i - 1][num_end + 1])
      if i + 1 < len(matrix):
        has_symbol |= is_symbol(matrix[i + 1][num_end + 1])
    if has_symbol:
      nums.append((int(matrix[i][num_start:num_end+1]), i, num_start, num_end))


  for j in range(len(matrix[i])):
    if matrix[i][j].isnumeric():
      num_end = j
      if i > 0:
        has_symbol |= is_symbol(matrix[i - 1][j])
      if i + 1 < len(matrix):
        has_symbol |= is_symbol(matrix[i + 1][j])
      if num_start == None:
        num_start = num_end
        if j > 0:
          has_symbol |= is_symbol(matrix[i][j - 1])
          if i > 0:
            has_symbol |= is_symbol(matrix[i - 1][j - 1])
          if i + 1 < len(matrix):
            has_symbol |= is_symbol(matrix[i + 1][j - 1])

    else:
      update_final(has_symbol)
      has_symbol = False
      num_start = num_end = None
  update_final(has_symbol)

nums
```

```python
gear_ratios = []

def get_num(i, j):
  values_in_row = [x for x in nums if x[1] == i]
  for (value, i, ns, ne) in values_in_row:
    if ns <= j <= ne:
      s = set()
      s.add((value, i, ns, ne))
      return s
  return set()

for i in range(len(matrix)):
  for j in range(len(matrix[i])):
    if matrix[i][j] != '*':
      continue
    found_nums = set()
    found_nums |= get_num(i - 1, j - 1)
    found_nums |= get_num(i - 1, j)
    found_nums |= get_num(i - 1, j + 1)
    found_nums |= get_num(i, j - 1)
    found_nums |= get_num(i, j + 1)
    found_nums |= get_num(i + 1, j - 1)
    found_nums |= get_num(i + 1, j)
    found_nums |= get_num(i + 1, j + 1)
    if len(found_nums) == 2:
      actual_nums = list(found_nums)
      gear_ratios.append(actual_nums[0][0] * actual_nums[1][0])

gear_ratios

sum(gear_ratios)
```
